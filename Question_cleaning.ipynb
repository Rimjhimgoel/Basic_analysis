```python
# Initialize empty lists to store results
section_results = []
question_results = []
time_taken_results = []

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    sections = row['sections']
    
    # Iterate over sections
    for section in sections:
        questions = section['questions']
        
        # Iterate over questions in the section
        for question in questions:
            question_id = question['questionId']  # Extract '_id' instead of 'question'
            time_taken = question['timeTaken']
            
            # Append to results lists
            section_results.append(section['_id'])  # Assuming you want to keep the entire section for reference
            question_results.append(question_id)
            time_taken_results.append(time_taken)

# Create a new DataFrame with extracted data
result_df = pd.DataFrame({
    'Section': section_results,
    'Question ID': question_results,  # Renamed to 'Question ID' to reflect '_id' field
    'Time Taken': time_taken_results
})

print(result_df)
```

```python
rows = []
other_columns = []

for index, row in df.iterrows():
    _id = row['_id']
    testId = row['testId']
    userId = row['userId']
    subjects = row['subjects']
    totalTimeTaken = row['totalTimeTaken']
    
    sections = row['sections']
    
    for section in sections:
        section_id = section['_id']
        questions = section['questions']
        
        for question in questions:
            question_id = question['questionId']
            time_taken = question['timeTaken']
            
            rows.append({
                '_id': _id,
                'testId': testId,
                'userId': userId,
                'Section': section_id,
                'Question ID': question_id,
                'Time Taken': time_taken,
                'subjects': subjects,
                'totalTimeTaken': totalTimeTaken
            })
            other_columns.append({
                '_id': _id,
                'testId': testId,
                'userId': userId,
                'subjects': subjects,
                'totalTimeTaken': totalTimeTaken
            })

expanded_df = pd.DataFrame(rows)

for col in ['_id', 'testId', 'userId', 'subjects', 'totalTimeTaken']:
    expanded_df[col] = [data[col] for data in other_columns]

print("\nExpanded DataFrame:")
print(expanded_df.head())
```
